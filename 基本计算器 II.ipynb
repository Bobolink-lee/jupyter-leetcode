```python
class Solution:

    '''
        解题思路：数据结构表达式计算（栈的使用）
            1、首先需要对所给字符串去空格，减少后续不必要的判断。
            2、声明两个栈，一个 操作数栈，一个 操作符栈。
                遍历字符串，遇到连续数字整合成一个整数，加入到操作数栈中。
                遇到 操作符o1 需要与操作符栈顶的 操作符元素o2 判断优先级，
                若 o2 是乘除运算符则其优先级最大；若 o2 是加减，o1 也是加减，则仍是o2大；
                若 o2 是加减，o1 是乘除则 o1 大。
                若是 o2 优先级大，则 o2 出栈，与操作数栈中的两个栈顶元素a,b进行计算 a o2 b，所得数再入回操作数栈中，继续判断当前操作符优先级；
                若是 o1 优先级大，则 o1 入栈 判断下一个字符；
            3、字符串已经遍历完成后，操作符栈未空，则说明计算未结束，继续做操作符出栈计算操作，直至为空，最后结果为操作数栈中的唯一元素。
    '''
    def calculate(self, s: str) -> int:
        # 字符串去空格
        s = s.replace(" ","")
        print(s)

        # 操作数栈 nums，操作符栈 oper
        nums = []
        oper = []

        # 遍历字符串
        i = 0
        while i < len(s):
            # 遇到连续数字，整合为一个整数
            if s[i] >= '0' and s[i] <= '9':
                total_num = 0
                while i < len(s) and s[i] >= '0' and s[i] <= '9':
                    total_num = total_num*10 + int(s[i])
                    i += 1
                nums.append(total_num)
                i -= 1
            # 遇到运算符，按情况操作
            else:
                # 操作符栈为空则可直接入栈
                if len(oper) == 0:
                    oper.append(s[i])
                # 否则视情况计算
                else:
                    # 操作符栈顶元素
                    o = oper[-1]
                    # 判断新的操作符o1优先级是否比操作符栈顶元素o2优先级大，是则入栈，否则计算
                    if (s[i] == '*' or s[i] == '/') and (o == '+' or o == '-'):
                        oper.append(s[i])
                    else:
                        oper.pop()
                        
                        # 取出操作数栈两个栈顶元素
                        b = nums.pop()
                        a = nums.pop()
                    
                        # 根据符号进行计算
                        if o == '*':
                            nums.append(a*b)
                        elif o == '/':
                            nums.append(a//b)
                        elif o == '-':
                            nums.append(a-b)
                        elif o == '+':
                            nums.append(a+b)
                        
                        # 下标回退，继续判断操作符优先级
                        i -= 1
            i += 1
            print("nums：",nums)
            print("oper：",oper)
        
        # 操作符栈未空，需要继续计算
        while len(oper) != 0:
            op = oper.pop()
            b = nums.pop()
            a = nums.pop()
                    
            if op == '*':
                nums.append(a*b)
            elif op == '/':
                nums.append(a//b)
            elif op == '-':
                nums.append(a-b)
            elif op == '+':
                nums.append(a+b)
            
            print("nums：",nums)
            print("oper：",oper)
        
        return nums[0]
    
    '''
        大佬思路：https://leetcode.cn/problems/basic-calculator-ii/solutions/648647/ji-ben-ji-suan-qi-ii-by-leetcode-solutio-cm28/
    '''
    def calculate1(self, s: str) -> int:
        
        # 单栈
        exp = []

        # 当前操作数
        num = 0
        # 上一个操作符，对于字符串中第一个元素可以视为其前一个操作符默认为 +
        pre_sign = '+'
        # 给字符串后尾加上个字符，减少不必要的判断
        s += '+'

        # 遍历字符串
        for temp in s:
            # 遇到空格跳过
            if temp == ' ':
                continue
            # 遇到数字整合为一个整数
            elif temp>='0' and temp<='9':
                if num:
                    num = num*10 + int(temp)
                else:
                    num = int(temp)
            # 遇到操作符，进行计算
            else:
                # 说明数据完整
                # 遇到 + ，则操作数直接入栈
                if pre_sign == '+':
                    exp.append(num)
                # 遇到 + ，则操作数取负号入栈
                elif pre_sign == '-':
                    exp.append(-num)
                # 遇到 * ，则操作数栈出栈与当前操作数进行计算后再入栈
                elif pre_sign == '*':
                    pre_num = exp.pop()
                    exp.append(pre_num*num)
                # 遇到 / ，则操作数栈出栈（需要取绝对值）与当前操作数进行计算后再入栈
                else:
                    pre_num = exp.pop()
                    new_num = int(abs(pre_num)//num)
                    if pre_num >= 0:
                        exp.append(new_num)
                    else:
                        exp.append(-new_num)
                # 重置当前操作数
                num = 0
                # 更新上一个操作符
                pre_sign = temp

        res = 0 # 最后一个数
        # 由于栈中已经处理了乘除运算，仅剩加减运算，且减运算已经视为操作数取负号形式，故可以直接累加即可
        for temp in exp:
            res += temp

        return res
```
